```python
import json
import pandas as pd
```

```python
population = pd.read_csv("input/population.csv")
life_expectancy = pd.read_csv("input/life-expectancy.csv")
gdp = pd.read_csv("input/gdp-per-capita.csv")
continents = pd.read_csv("input/continents-according-to-our-world-in-data.csv")
```

```python
countries_data = []

for idx, row in life_expectancy.iterrows():
    obj = {}
    countryPopulation = population[(population['Entity'] == row['Entity']) & (population['Year'] == row['Year'])]
    countryGDP = gdp[(gdp['Entity'] == row['Entity']) & (gdp['Year'] == row['Year'])]
    countryContinent = continents[(continents['Entity'] == row['Entity'])]
    
    if ((countryPopulation.shape[0] == 1) & (countryGDP.shape[0] == 1) & (countryContinent.shape[0] == 1)):
        obj['country'] = row['Entity']
        obj['year'] = int(row['Year'])
        obj['life_expectancy'] = row['Life expectancy']
        obj['population'] = int(countryPopulation.iloc[0]['Population (historical estimates)'])
        obj['gdp'] = countryGDP.iloc[0]['GDP per capita']
        obj['continent'] = countryContinent.iloc[0]['Continent']
        
        countries_data.append(obj)
```

```python
countries_data
```

```python
with open("output/rosling.json", "w") as json_file:
    this_file = json.dumps(countries_data, ensure_ascii=False)
    json_file.write(this_file)
```
